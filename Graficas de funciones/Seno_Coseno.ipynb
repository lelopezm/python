# Gráfica de Seno y Coseno

En este código vamos a graficar $f(x)=\sin{x}$ y $g(x)=\cos{x}$ en $x\in[0,2\pi]$, identificando sus puntos de intersección entre ellas.

```python
import numpy as np #libreria numérica de python
import matplotlib.pyplot as plt #libreria de gráficos

def f(x):
  y=np.sin(x)
  return y

def g(x):
  y=np.cos(x)
  return y

#dominio [x0,xf]
x0=0
xf=2*np.pi
X=np.linspace(x0,xf,500)
Y1=f(X) #función seno
Y2=g(X) #función coseno
plt.plot(X,Y1,'#0B0FF2',linewidth=1.5,label='$f(x)=sen(x)$')
plt.plot(X,Y2,'#F2890B',linewidth=1.5,label='$f(x)=cos(x)$')
plt.scatter(np.pi/4,f(np.pi/4),color='#1CF20B',linewidths=2.0, zorder=5)
plt.scatter(5*np.pi/4,f(5*np.pi/4),color='#F20B35',linewidths=2.0, zorder=5)
plt.text(0,0,'$0$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
plt.text(np.pi/2,0,'$\dfrac{\pi}{2}$', color='black',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(np.pi,0,'$\pi$', color='black',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(3*np.pi/2,0,'$\dfrac{3}{2}\pi$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
plt.text(2*np.pi,0,'$2\pi$', color='black',fontsize=10, ha='left',va='top', fontweight='bold')
#puntos de intersección
plt.text(np.pi/4,0,'$\dfrac{\pi}{4}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(0,np.sqrt(2)/2,'$\dfrac{\sqrt{2}}{2}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(5*np.pi/4,0,'$\dfrac{5}{4}\pi$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.text(0,-np.sqrt(2)/2,'$\dfrac{-\sqrt{2}}{2}$', color='red',fontsize=10, ha='right',va='top', fontweight='bold')
plt.plot([np.pi/4,np.pi/4],[0,np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([0,np.pi/4],[np.sqrt(2)/2,np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([5*np.pi/4,5*np.pi/4],[0,-np.sqrt(2)/2],'--r',linewidth=0.7)
plt.plot([0,5*np.pi/4],[-np.sqrt(2)/2,-np.sqrt(2)/2],'--r',linewidth=0.7)
#ejes coordenados
plt.axhline(0, color='black', linewidth=1.5) #eje x
plt.text(xf+0.3,0,'$x$', fontsize=10, ha='left', va='top') #etiqueta del eje x
plt.axvline(0, color='black', linewidth=1.5) #eje y
plt.text(0,1,'$y$', fontsize=10, ha='right', va='bottom') #etiqueta del eje x
plt.grid() #mallado
plt.xticks(np.arange(x0,xf,np.pi/2)) #control mallado horizontal
plt.legend()
plt.show()
```
