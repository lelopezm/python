# Crecimiento exponencial

Ley de Malthus:
$$P(t)=P_0 e^{kt}$$
donde $P_0$ es la población inicial, $k$ es la tasa de crecimiento de la población, $t$ el tiempo (variable independiente) y $P$ es el tamaño promedio de la población (variable dependiente).

```python
import numpy as np #libreria numérica de python
import matplotlib.pyplot as plt #libreria de gráficos

#Definimos la función a graficar
def P(t):
    return P0*np.exp(k*t)

#definimos el tiempo en horas
t0=0
tf=5

# Valores parámetros
P0=1300 #población inicial
te=1 #al cabo de una hora
Pe=2*P0 #población en tiempo te
k=(1/te)*np.log(Pe/P0)

T=np.linspace(t0,tf,500) #linea de tiempo
Pt=P(T)
plt.plot(T,Pt,'#EAD51B',linewidth=1.5)
plt.scatter(0,P0,color='#EA631B',linewidths=2.0, zorder=5) #marcamos la condición inicial
plt.scatter(te,Pe,color='#EA631B',linewidths=2.0, zorder=5) #marcamos la condición observada (te,Pe)
plt.axhline(0, color='black', linewidth=1) #eje T
plt.text(tf, -0.1, '$t$', fontsize=10, ha='left', va='top') #etiqueta del eje t
plt.axvline(0, color='black', linewidth=1) #eje P
plt.text(0, max(Pt), '$P(t)$', color="black",fontsize=10, ha='left', va='bottom') #etiqueta del eje P
plt.grid() #mallado
plt.xlabel('tiempo (horas)')
plt.ylabel('Número de bacterias')
plt.savefig('bacterias.pdf',bbox_inches='tight',pad_inches=0) #para guardar la grafica
plt.show()
```

```python
# calcular el tiempo para que la población sea Pa
Pa=10000
ta=(1/k)*np.log(Pa/P0)
decimalta, enterata=np.modf(ta)
minta=decimalta*60
decimalminta, enteraminta=np.modf(minta)
segta=decimalminta*60
decimalsegta, enterasegta=np.modf(segta)
print('La población será de %1.1f bacterias al cabo de %i horas, %i minutos, %i segundos, aproximadamente.'%(Pa,enterata,enteraminta,enterasegta))
```
