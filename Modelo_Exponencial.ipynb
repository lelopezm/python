# Crecimiento exponencial

Ley de Malthus:
$$P(t)=P_0 e^{kt}$$
donde $P_0$ es la población inicial, $k$ es la tasa de crecimiento de la población, $t$ el tiempo (variable independiente) y $P$ es el tamaño promedio de la población (variable dependiente).

```python
import numpy as np #libreria numérica de python
import matplotlib.pyplot as plt #libreria de gráficos

#Definimos la función a graficar
def P(t):
    return P0*np.exp(k*t)

#definimos el tiempo 
t0=0
tf=3

# Valores de los coeficientes y=ax^2+bx+c
P0=5
k=1.15

T=np.linspace(t0,tf,500) #linea de tiempo
Pt=P(T)
plt.plot(T,Pt,'#EAD51B',linewidth=1.5)
plt.scatter(0,P0,color='#EA631B',linewidths=2.0, zorder=5) #marcamos la condición inicila
#plt.text(vx, f(vx)+0.2, '$V$', color="red",fontsize=12, ha='center', va='bottom',fontweight='bold') #etiqueta del vértice
plt.axhline(0, color='black', linewidth=1) #eje T
plt.text(tf, -0.1, '$t$', fontsize=10, ha='left', va='top') #etiqueta del eje t
plt.axvline(0, color='black', linewidth=1) #eje P
plt.text(0, max(Pt), '$P(t)$', color="black",fontsize=10, ha='center', va='bottom') #etiqueta del eje P
plt.grid() #mallado
#plt.xticks(np.arange(-1,5,0.5)) #control mallado horizontal
#plt.yticks(np.arange(-6,5,1)) #control mallado vertical
plt.xlabel('tiempo')
plt.ylabel('población')
#plt.savefig('parabola.pdf',bbox_inches='tight',pad_inches=0) #para guardar la grafica
plt.show()
```

```python
# calcular el tiempo para que la población sea Pa
Pa=80
ta=(1/k)*np.log(Pa/P0)
print('La población será de %1.1f individuos al cabo de %1.2f días.'%(Pa,ta))
```
